```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the datasets
gdp_per_capita_path = 'data/US Real gross domestic product per capita.csv'
income_per_capita_path = 'data/US Real net national disposable income per capita CVM SA.csv'

gdp_per_capita_df = pd.read_csv(gdp_per_capita_path)
income_per_capita_df = pd.read_csv(income_per_capita_path)

# Display the first few rows of each dataset to understand their structure
(gdp_per_capita_df.head(), income_per_capita_df.head())

# Filter the datasets to include only data from 2010 onwards
gdp_per_capita_df['DATE'] = pd.to_datetime(gdp_per_capita_df['DATE'])
income_per_capita_df['DATE'] = pd.to_datetime(income_per_capita_df['DATE'])

gdp_filtered = gdp_per_capita_df[gdp_per_capita_df['DATE'].dt.year >= 2010]
income_filtered = income_per_capita_df[income_per_capita_df['DATE'].dt.year >= 2010]

# Plotting
plt.figure(figsize=(12, 7))
plt.plot(gdp_filtered['DATE'], gdp_filtered['A939RX0Q048SBEA'], label='GDP per Capita', marker='x', linestyle='--', color='red')
plt.plot(income_filtered['DATE'], income_filtered['A229RX0'], label='Income per Capita', marker='o' ,linestyle='-', color='green')

# Formatting the plot
plt.title('US GDP per Capita vs. Income per Capita (2010 Onwards)')
plt.xlabel('Year')
plt.ylabel('Value in USD')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Show plot
plt.show()
```
